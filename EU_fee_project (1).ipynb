Abstract

*This paper aims to understand the impact of tuition fee increases in England on EU international undergraduate student mobility. First, we compared changes in EU undergraduate enrolments in England, which trebled its tuition fee levels in 2012/13 from £3,375 to £9,000, with those observed in Scotland, which does not charge tuition fees to EU undergraduate students. Second, we look at the changes in enrolment by field of study and type of higher education institution by EU countries. Our analysis shows that, on average, English universities suffered a sudden drop in EU enrolments in 2012/13 but recovered their pre-2012 enrolment levels in subsequent academic years. We observe that those English universities regarded as less prestigious experienced the sharpest decline and took longer to recover their pre-2012 numbers. Still, we found different student mobility patterns when examining differences between EU countries. While there is a significant decrease in the number of students from some Western European countries (mainly German and French students) attending English universities after 2012, the tuition fees increase did not impact the number of students coming from Southern Europe. Moreover, we found country-specific patterns regarding the field of studies: a high number of students pursuing subjects allied to Medicine from Ireland, Creative Arts and Design from Portugal, Social Studies from Italy, and Computer Sciences from Romania. Our preliminary results support research on EU intra-mobility that indicates that mobility can be an alternative route of capital accumulation for students from more marginal countries.*

## 2. Literature Review

### 2.1 The impact of tuition fee changes on student enrolments

In recent decades, there has been a substantial increase in academic attention to higher education funding regimes and their impact on student behaviour. This has coincided with changes in the political economy of higher education finance brought about by austerity drives, the growing costs of expanding higher education systems and the emergence and consolidation of neoliberal economic discourses. The latter seek to model ‘the world in terms of functioning capitalist markets [and] its policy objective is to reform institutions, systems, subjects and behaviour to render them instrumental for capital accumulation’ (Marginson 2013, p. 354) and where higher education is perceived ‘as a private good, which benefits individuals, and which individuals should therefore pay, rather than a public good, which benefits societies and economies’ (Wilkins, Shams and Huisman 2013, p. 125). Inter alia, governments inspired by neoliberal policy frameworks have progressively shifted the cost of higher education from the state to the student (Miller 2010) and have favoured targeted rather than universal subsidies to address inequities in access (Leslie and Brinkman 1987, Shattock 2012). In the following lines, we review the impact that these changes have had on different national contexts and to different pools of students. We will also pay particular attention to existing research that has looked at the impact of tuition fee changes on internationally mobile students.


First, it is important to highlight that existing research on student finance shows that there are substantial differences between countries –and within countries– in the tuition fee levels that students need to pay and, equally important, the subsidies –or loans– students may have access to while in higher education. In this sense, countries’ tuition-subsidy arrangements ‘have been found to be of enormous importance because of their far-reaching social, economic and political consequences: the tuition-subsidy system affects students’ enrolment behaviour, and thereby shape, among other things, patterns of social mobility and socio-economic as well as educational inequality’ (Garritzmann 2016, p. 301-302). These cross-country differences in tuition-subsidy regimes might be the reason why evidence about the effect of tuition fee changes and increases on shaping students’ decision-making processes is mixed, particularly because some of the potential negative effects of tuition fee changes on participation cannot be disentangled from the ameliorating effects of support packages, including student loans.
In the United States, Leslie and Brinkman (1987) found, after reviewing 25 studies on the effects of tuition fees and financial aid on student enrolment, that an increase of $100 in tuition fees ‘one would expect an 18-24-year-old participation rate drop of about three-quarters of a percentage point’ (p. 188-189, emphasis in the original). In their discussion, they also suggest that increases in tuition fee prices appear to have a greater effect that other price changes, such as accommodation or travel, and they offer an explanation as to why this may be the case:</div>

<div align="right">‘Several explanations have been offered as to why tuition response might be relatively large. Tuition is the most visible college price, and it is the one that is most inescapable. College tuitions are conspicuous, and students are unusually conscious of them. Annual increases generally are well publicized and often are debated publicly’(p. 196).</div>

Ten years later, Heller (1997) updated Leslie and Brinkman’s (1987) study. Again, they analysed a corpus of research investigating the relationship between financial availability and access to higher education, in this case with up-to-date data and putting special emphasis on price and financial aid sensitivity among students from different incomes, ethnicities and institutional types. They found that, indeed, there is ‘a downward-sloping demand curve for higher education’ (Heller 1997, p. 650). In particular, they showed that increases in tuition levels and declines in financial aid –particularly grants– yield lower enrolment. Moreover, they found that students from low income households, black students, and those studying in community colleges are more sensitive to tuition changes. Similarly, Hübner (2012) found that this relationship exists among German prospective undergraduate students. Using a natural experimental design, they compared enrolment changes between German states that introduced tuition fees in 2007 and those that did not, concluding that ‘the introduction of tuition fees at an annual rate of €1,000 reduced enrolment probabilities amongst high-school graduates from states which introduced fees by 2.7 percentage points’ (p.959).


In the case of Australia, Foskett, Roberts and Maringe (2006) found a radically different picture, probably due to the nature of the changes in student finance put in place in these two countries. They analysed the impact of the Higher Education Contribution Scheme (HECS), introduced in 1989. HECS is a fee charged to all undergraduate students the payment of which can be deferred until graduates reached a certain level of income, ‘this was the world’s first income-contingent charge for higher education’ (p. 31). According to the authors, evidence from Australia does not suggest that ‘increasing fees reduces participation in HE’ (p. 110). However, there is one important issue to consider from the Australian case. The conditions under which Australian undergraduate students borrow resources to pay for their higher education are more attractive than in other high-fee regimes, and borrowing levels appear to be significantly lower than, for instance, the United States. In the latter, ‘even six-digit debts are not uncommon’ (Garritzmann 2016, p. 160). This suggests that, under certain circumstances such as when government-backed income-contingent loans are available, access to credit rather than debt aversion may be key to understand participation (Wakeling et al., 2017; Mateos-González and Wakeling, in press).


In the case of England, this paper’s country of interest, research on the potential impact of tuition fee changes on enrolment mushroomed since the introduction of variable fees of up to £3,000 in 2006 (Foskett, Roberts and Maringe, 2006) and got momentum when it was announced, in 2010, that the cap on tuition fees would be raised to £9,000 starting in 2012 –the piece of policy being investigated in this article. This decision was indeed controversial, triggering student protests across the country (Davis, Cecil, Murphy and Davenport 2010). One of the first pieces of research looking at the impact of this policy was produced by Wilkins, Shams and Huisman (2013), who surveyed 1549 high school students about to make the decision to apply to university when the £9,000-fee policy was introduced. The researchers concluded that financial issues were ‘the most important factor’ for students’ study choices, ‘with a quarter of respondents considering postponing university studies and almost one-fifth considering cheaper higher education options’ (p. 136). Wilkins, Shams and Huisman (2013) conclusion is backed up by later research. Geven (2015), using a quasi-experiment design comparing enrolment rates between students who did not have to pay fees in 2012 in the UK –e.g. Scottish students– and those who did, found that ‘enrolment declined by 15\% in treated groups as a result of the tuition fee increase’ (p. 479). According to Geven (2015), this effect was particularly important among older students and, interestingly, for those from ‘the middle and service class’ (p. 495). Similarly. Sá (2014) found that the £9,000-fee policy has a smaller effect on ‘non-white students and for students from local authorities with low rates of participation in higher education’ (p. 2), and she offers a potential explanation for why this might be the case: that fees are subsidies via government-backed income-contingent loans and that, under the current arrangements, ‘universities that charge higher fees need to have access agreements […] to attract and support students from disadvantaged backgrounds’ (ibid.).


## 3. Data

We use the Higher Education Statistics Agency (HESA) data for first-year EU undergraduate students that attended a UK higher education institution between the academic years 2007/08 and 2018/19. The tables and graphs present in this paper follow the HESA standard rounding methodology  which prevents identifying individuals in the data.
In this paper, we seek to capture the effect of changes in the cost of higher education in England on EU student enrolments in English higher education institutions. By using a quasi-experimental method, we take advantage of the variation in tuition fee regimes in UK home nations and compare changes in enrolment rates in England, where there was a threefold increase in undergraduate tuition fees in 2012, to those in Scotland, which has not charged tuition fees to EU undergraduate students throughout the period under study.
